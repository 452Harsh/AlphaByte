```python
import os
from dotenv import load_dotenv
import google.generativeai as genai

# Load environment variables from the .env file
load_dotenv()

# Configure the API key for Google Generative AI
genai.configure(api_key=os.getenv("GEMINI_SECRET_KEY"))

class GenerativeAIModel:
    _instance = None

    def __new__(cls):
        if not cls._instance:
            cls._instance = super().__new__(cls)
            cls._instance.model = genai.GenerativeModel("gemini-1.5-flash")
        return cls._instance

def translate_and_analyze_kannada_text(kannada_text):
    """
    Translate the Kannada transcription about sandalwood cultivation and extract key 
    information related to cultivation practices, care techniques, and environmental conditions.

    Parameters:
        kannada_text (str): The Kannada text to translate and analyze.
    
    Returns:
        str: The translated content in English, highlighting important practices for sandalwood cultivation.
    """
    try:
        # Initialize model
        model = GenerativeAIModel()._instance.model
        
        # Generate translation and insights, focusing on relevant cultivation information
        prompt = (
            f"Translate the following Kannada text to English. Provide key insights about "
            f"sandalwood cultivation, such as care techniques, seasonal requirements, environmental "
            f"conditions, and any specific practices mentioned in the text.\n\n{kannada_text}\n\n"
            f"Provide clear, concise, and actionable information."
        )
        
        response = model.generate_content(prompt)
        
        # Check if the model's response is valid
        if response.text.strip():
            return response.text.strip()
        else:
            # Return a default message if the response is empty
            return "Translation or insights could not be generated."

    except Exception as e:
        print("An error occurred:", e)
        # Return a message indicating the failure
        return f"Error during translation: {str(e)}"

def process_all_txt_files(input_folder, output_folder):
    """
    Processes all .txt files in the input folder, translates and analyzes their content, 
    and saves the output to the output folder.

    Parameters:
        input_folder (str): The path to the folder containing .txt files with Kannada text.
        output_folder (str): The path to the folder where translated files with insights will be saved.
    """
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Get a sorted list of .txt files to process sequentially
    txt_files = sorted([f for f in os.listdir(input_folder) if f.endswith(".txt")])

    # Iterate over all .txt files in the sorted order
    for filename in txt_files:
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, f"translated_{filename}")

        print(f"Processing file: {filename}")

        # Read the Kannada text from the file
        with open(input_path, "r", encoding="utf-8") as file:
            kannada_text = file.read()

        # Translate and analyze the Kannada text
        translated_text = translate_and_analyze_kannada_text(kannada_text)

        # Debugging: Print the translated text to check if it's None or empty
        print(f"Translated text for {filename}: {translated_text}")

        # Check if translated text is None or contains an error message
        if translated_text is None or "Error" in translated_text:
            print(f"Failed to translate {filename}. Skipping...\n")
            continue

        # Save the translated text with insights to the output file
        if translated_text:  # Ensure that the translated text is not None or empty
            with open(output_path, "w", encoding="utf-8") as file:
                file.write(translated_text)
            print(f"Translated content for {filename} has been saved to {output_path}\n")
        else:
            print(f"Skipping {filename} because the translation is empty or invalid.\n")

# Set your folder paths
input_folder = "/Users/harshtuli/Desktop/Whishper-venv/output"
output_folder = "/Users/harshtuli/Desktop/Whishper-venv/translate"

# Run the translation and analysis
process_all_txt_files(input_folder, output_folder)
```

```python
import os
import json

# Define the folder path containing the text files and output file path
translate_folder = "/Users/harshtuli/Desktop/Whishper-venv/translate"  # Update with your actual folder path
output_file = "sandalwood_context_dataset.json"

# Initialize list to hold each file's context
context_dataset = []

# Process each file in the folder
for idx, filename in enumerate(os.listdir(translate_folder)):
    if filename.endswith(".txt"):
        file_path = os.path.join(translate_folder, filename)
        
        with open(file_path, 'r') as file:
            content = file.read()
            
            # Create a dictionary for the file content
            file_entry = {
                "file_id": f"file_{idx+1:02d}",  # Assign a unique ID for each file
                "context": content.strip()  # Strip any extra whitespace
            }
            
            # Append the entry to the dataset list
            context_dataset.append(file_entry)

# Save the dataset to JSON file
with open(output_file, 'w') as f:
    json.dump(context_dataset, f, indent=2)

print(f"Dataset created with {len(context_dataset)} entries and saved to {output_file}")
```

```python
import json
from transformers import pipeline

# Load the dataset
with open("/Users/harshtuli/Desktop/Whishper-venv/myenv/sandalwood_context_dataset.json", 'r') as f:
    data = json.load(f)

# Initialize a pre-trained Question Answering pipeline
qa_pipeline = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")

# Function to find the most relevant context and answer a question
def answer_question(question, data):
    # Step 1: Retrieve relevant contexts (simplified keyword matching example)
    relevant_contexts = []
    for entry in data:
        if any(keyword in entry["context"].lower() for keyword in question.lower().split()):
            relevant_contexts.append(entry)
    
    # Step 2: Run QA model on relevant contexts and return the best answer
    best_answer = None
    best_score = 0
    
    for entry in relevant_contexts:
        context = entry["context"]
        result = qa_pipeline({"question": question, "context": context})
        if result["score"] > best_score:
            best_answer = result["answer"]
            best_score = result["score"]
    
    return best_answer if best_answer else "No relevant information found."

# Example usage
question = "What is the best soil type for sandalwood?"
answer = answer_question(question, data)
print(f"Answer: {answer}")
```

```python
import os
import json
from dotenv import load_dotenv
import google.generativeai as genai

# Load environment variables from the .env file
load_dotenv()

# Configure the API key for Google Generative AI
genai.configure(api_key=os.getenv("GEMINI_SECRET_KEY"))

class GenerativeAIModel:
    _instance = None

    def __new__(cls):
        if not cls._instance:
            cls._instance = super().__new__(cls)
            cls._instance.model = genai.GenerativeModel("gemini-1.5-pro")
        return cls._instance

def generate_summary(text):
    """
    Generate a summary of the given text using the Gemini 1.5 model.
    
    Parameters:
        text (str): The text to summarize.
    
    Returns:
        str: The generated summary.
    """
    try:
        # Initialize model
        model = GenerativeAIModel()._instance.model
        
        # Create the prompt for summarization
        prompt = (
            f"Summarize the following text about sandalwood cultivation. Provide a concise, actionable summary:\n\n{text}\n\n"
            "The summary should include key points related to care techniques, environmental conditions, and best practices."
        )
        
        # Generate the summary
        response = model.generate_content(prompt)
        
        # Return the summary text, or handle an empty response
        return response.text.strip() if response.text.strip() else "Summary could not be generated."
    
    except Exception as e:
        print("An error occurred:", e)
        return f"Error during summary generation: {str(e)}"

def summarize_json(input_json_path, output_json_path):
    """
    Summarizes the content of each entry in the JSON file and writes the summarized data to a new JSON file.
    
    Parameters:
        input_json_path (str): The path to the input JSON file.
        output_json_path (str): The path to the output JSON file where summaries will be saved.
    """
    # Load the input JSON data
    with open(input_json_path, 'r', encoding='utf-8') as infile:
        data = json.load(infile)
    
    summarized_data = []

    # Iterate through each entry in the JSON data
    for entry in data:
        file_id = entry.get('file_id')
        context = entry.get('context')

        if not file_id or not context:
            print(f"Skipping entry due to missing 'file_id' or 'context': {entry}")
            continue

        print(f"Generating summary for file_id: {file_id}")

        # Generate the summary for the context
        summary = generate_summary(context)

        # Append the summarized result to the list
        summarized_data.append({
            'file_id': file_id,
            'summary': summary
        })

        print(f"Summary for file_id {file_id}: {summary}\n")

    # Save the summarized data to the output JSON file
    with open(output_json_path, 'w', encoding='utf-8') as outfile:
        json.dump(summarized_data, outfile, indent=4)

    print(f"Summarized data saved to {output_json_path}")

# Set your input and output paths
input_json_path = "/Users/harshtuli/Desktop/Whishper-venv/myenv/sandalwood_context_dataset.json"  # Replace with your actual path
output_json_path = "/Users/harshtuli/Desktop/Whishper-venv/myenv/output_summaries.json"  # Replace with your actual path

# Run the summarization
summarize_json(input_json_path, output_json_path)
```
